# IRIS CLASS CLASSIFICATION

- A machine learning project using Python. 
- This project is a supervised learning, classification problem with multiple classes.
- In this project, we will compare and contrast the performance of three different learning models on a common and publicly       available dataset.

- We will be using the iris flowers dataset, which contains 150 observations of iris flowers.
- There are four columns of measurements and the species of flower observed.
- Only three species are present in this dataset.


### Importing the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv("C:\\Users\\SAMEER\\Desktop\\IRIS.csv")
```

```python
dataset.head()
```

### Visualising the dataset

```python
dataset.hist(figsize=(10, 5))
plt.show()
```

```python
# scatter plot matrix
from pandas.plotting import scatter_matrix
scatter_matrix(dataset)
plt.figure(figsize=(30,20))
plt.show()
```

### Correlation matrix 

```python
sns.heatmap(dataset.corr(), annot = True)
```

```python
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, 4]
```

```python
X
```

```python
Y
```

### Preprocessing the dataset

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
Y = labelencoder.fit_transform(Y)
```

```python
Y
```

### Spliting the dataset into train and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 123)
```

### Applying the Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
```

```python
X_train
```

## Applying the Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier_logistic= LogisticRegression(random_state =0)
classifier_logistic.fit(X_train, Y_train)
```

```python
Y_pred_logistic = classifier_logistic.predict(X_test)
```

```python
Y_pred_logistic
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
cm_logistic = confusion_matrix(Y_test, Y_pred_logistic)
```

```python
sns.heatmap(cm_logistic, annot = True, fmt = 'g')
```

```python
accuracy_score(Y_test, Y_pred_logistic)
```

```python
print(classification_report(Y_test, Y_pred_logistic))
```

## Applying the RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
cm_rand = confusion_matrix(Y_test, Y_pred)
```

```python
sns.heatmap(cm_rand, annot=True, fmt = 'g')
```

```python
accuracy_score(Y_test, Y_pred)
```

```python
print(classification_report(Y_test, Y_pred))
```
